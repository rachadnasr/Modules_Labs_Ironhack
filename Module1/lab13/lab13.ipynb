```python
#4. Add username and password as parameters
ploads = {"username": "u", "password": "p"}
response = requests.get("https://httpbin.org/get", params = ploads)
results = response.json()
results
```

## Challenge 3: Hidden Cold Joke

```python
#search for these files by calling the Github API
response = requests.get('https://api.github.com/search/code?q=.scavengerhunt+in:path+repo:ironhack-datalabs/scavenger')
search = response.json()
search
```

```python
#1. Sort the filenames ascendingly.
search["items"]
repo_files_names = [item["name"] for item in search["items"]]
repo_files_names.sort()
repo_files_names
```

```python
#2. Read the content of each secret files into an array of strings.
repo_files_urls = {item["name"]: item["url"] for item in search["items"]}
repo_files_urls
```

```python
repo_files = [requests.get(repo_files_urls[repo_filename]) for repo_filename in repo_files_names]
repo_files
```

```python
repo_files_content = [file.json()["content"] for file in repo_files]
repo_files_content
```

```python
import base64
message = [base64.b64decode(b64message).decode('ascii') for b64message in repo_files_content]
message
```

```python
#3. Concatenate the strings in the array separating each two with a whitespace.
#4. Print out the joke.
cold_joke = " ".join([str(i) for i in message])
cold_joke = cold_joke.replace("\n", "")
cold_joke
```
