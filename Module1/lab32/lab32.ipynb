### Ordinal Encoding 

#### with python librairie

```python
data = {
    'TransactionID': [1, 2, 3, 4, 5],
    'ClientID': [231, 765, 453, 231, 892],
    'Profession': ['self-employed', 'students', 'Horeca', 'self-employed', 'finance'],
    'Bank_dep': ["009", "005", "007", "009", "003"],
    'Risk': ['high', 'high', 'medium', 'high', 'low'],
    'Number of credits': [2, 3, 5, 2, 3],
    'Revenue':[30200, 12700, 89400, 30200, 740000]
}

data = pd.DataFrame(data=data)
```

```python
#  assign the original order of the variable through a dictionary.
Profession_dict = {'students': 1,
                   'self-employed': 2,
                   'Horeca': 3,
                   'finance': 4}
Risk_dict = {'low': 1,
             'medium': 2,
             'high': 3}
Bank_dep_dict = {'003' : 1,
           '005':2,
           '007':3,
           '009':4}
```

```python
# Then we can map each row for the variable as per the dictionary.
data['Profession_ordinal'] = data.Profession.map(Profession_dict)
data['Bank_dep_ordinal'] = data.Bank_dep.map(Bank_dep_dict)
data['Risk_ordinal'] = data.Risk.map(Risk_dict)
data
```
